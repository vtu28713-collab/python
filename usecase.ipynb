```python
import matplotlib.pyplot as plt
import numpy as np

# Define parameters for binomial distribution
n = 10  # Number of trials
p = 0.5 # Probability of success

# Values for number of successes
x = np.arange(0, n+1)

# Calculate binomial probabilities
probabilities = [np.math.comb(n, k) * (p**k) * ((1-p)**(n-k)) for k in x]

# Plot the probabilities as a bar chart
plt.bar(x, probabilities, color='skyblue')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution Probabilities (n=10, p=0.5)')
plt.show()
```
